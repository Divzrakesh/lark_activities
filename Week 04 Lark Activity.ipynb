```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('Telco-Customer-Churn.xlsx')
```

```python
data
```

# 1. Compare churn count with respect to gender


```python
plt.figure(figsize=(10,6))
sns.countplot(x=data['gender'],data=data,hue ='Churn')
plt.title('Gender Vs Churn count')
plt.xlabel('Gender')
plt.ylabel('Churn Count')
plt.grid(linestyle='--')
plt.show()
```

# 2. Count of female senior citizens


```python
fsc = data[data['SeniorCitizen']==1].groupby(['gender','SeniorCitizen']).size().reset_index(name='count')
fsc
```

```python
plt.figure(figsize=(3,6))
fsc = data[(data['gender']=='Female')&(data['SeniorCitizen']==1)]
sns.countplot(x=fsc['gender'])
plt.title('Count of female senior citizens')
plt.show()
```

# 3. Tenure Vs Total Charges

```python
plt.figure(figsize=(10,6))
plt.scatter(data['tenure'],data['TotalCharges'],s=10,c= 'blue')
plt.title('Plot of Tenure vs Total Charges')
plt.xlabel('Tenure')
plt.ylabel('Total Charges')
plt.show()
```

# 4. Find out which contract is preferred by the senior citizen


```python
plt.figure(figsize=(10,6))
sc = data[data['SeniorCitizen']==1]
sns.countplot(x='Contract',data =sc)
plt.title('Preferred contract of Senior citizens')
plt.xlabel('Contract type')
plt.ylabel('Count')
plt.grid()
plt.show()
```
