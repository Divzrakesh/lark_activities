# 5. Find out which region has more consistent with respect to GDP per capita (current US$)


```python
df2=df.groupby('Region')[['GDP per capita (current US$)']].std()
df2.sort_values('GDP per capita (current US$)')
```

```python
df2.nsmallest(1,'GDP per capita (current US$)')
```
